```python
import pandas as pd
df = pd.read_csv(DATASET_PATH)
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.to_parquet("/content/cabai_merah_keriting.parquet")
```

```python
os.environ["GIFT_EVAL"] = "/content/cabai_merah_keriting.parquet"
```

```python
from gift.eval import zeroshot_eval

zeroshot_eval(
    dataset_name="/content/cabai_merah_keriting.parquet",  # Pakai dataset yang baru kita simpan
    context_length=96,  # Sesuaikan dengan dataset
    forecast_length=24,
    batch_size=4  # Jangan terlalu besar agar tidak kehabisan RAM
)
```
